```python
import json
import numpy as np
import pandas as pd
import datetime as dt
from pandas.io.json import json_normalize
np.set_printoptions(precision=4, suppress=True)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
fcast_data_ifp = pd.read_csv('data0928/fcasts0928_with_ifp_turk_info.csv')
```

```python
with open('data0928/forecasts.json') as json_file_2:
    fcast_data = json.load(json_file_2)
    
print(len(fcast_data))#, len(ifp_data))
fcast_data = fcast_data['predictions'] #'predictions' is the only first tier key of the json file
print(len(fcast_data))#, len(ifp_data))
fcasts = json_normalize(fcast_data)
fcasts = fcasts.sort_values(by=['question_id','user_id','created_at'], ascending=[True, True, True])
fcasts = fcasts.reset_index()

fcasts.columns
```

```python
fcast_data_ifp['forecast'] = fcasts['forecast']
```

```python
ifp_control_users = fcast_data_ifp[fcast_data_ifp['user_id'] > 9000]
ifp_control_users = ifp_control_users.reset_index()
```

```python
import time
import datetime
created_at = ifp_control_users['created_at'].astype(str).str[:10]
closes_at = ifp_control_users['ifp_closes_at']
resolves_at = ifp_control_users['ifp_resolves_at']
delta1s = []
delta2s = []
for k in range(len(created_at)):
    #print(k)
    created = created_at[k]
    closes = closes_at[k]
    resolves = resolves_at[k]
    #print(created)
    #print(closes)
    #print(resolves)
    delta1 = time.mktime(datetime.datetime.strptime(closes, "%Y-%m-%d").timetuple()) - time.mktime(datetime.datetime.strptime(created, "%Y-%m-%d").timetuple()) 
    delta2 = time.mktime(datetime.datetime.strptime(resolves, "%Y-%m-%d").timetuple()) - time.mktime(datetime.datetime.strptime(created, "%Y-%m-%d").timetuple()) 
    delta1s.append(delta1)
    delta2s.append(delta2)
    #print(timestamp1)
for k in range(len(delta1s)):
    delta1s[k] = delta1s[k] / 86400
    delta2s[k] = delta2s[k] / 86400
#fcast_data_ifp['time_elapsed'] = fcast_data_ifp['timestamp'] - fcast_data_ifp['created_at']
```
